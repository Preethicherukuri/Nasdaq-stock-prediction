# NASDAQ Composite Stock Market Prediction

```python
import yfinance as yf
```

```python
nasdaq = yf.Ticker("^IXIC")
```

```python
nasdaq = nasdaq.history(period="max")
```

```python
nasdaq.index
```

## Data Cleaning and Visualization of NASDAQ Stock Market

```python
nasdaq.plot.line(y="Open", use_index =True)
```

```python
# This column is more used for individual stock not in index
del nasdaq["Dividends"]
del nasdaq["Stock Splits"]
```

```python
nasdaq["Tomorrow"] = nasdaq["Open"].shift(-1)
```

```python
nasdaq
```

```python
nasdaq["Target"] = (nasdaq["Tomorrow"] > nasdaq["Open"]).astype(int)
```

```python
nasdaq
```

```python
nasdaq = nasdaq.loc["2000-01-01":].copy()
```

```python
nasdaq
```

# Training the model

```python
from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier(n_estimators=100, min_samples_split=100, random_state=1)

train = nasdaq.iloc[:-100]
test  = nasdaq.iloc[-100:]

predictors = ["Close", "Open", "Low", "High", "Volume"]
model.fit(train[predictors], train["Target"])
```

```python
from sklearn.metrics import precision_score

preds  = model.predict(test[predictors])
```

```python
preds
```

```python
import pandas as pd

preds = pd.Series(preds, index =  test.index)
```

```python
precision_score(test["Target"], preds)
```

```python
combined = pd.concat([test["Target"], preds], axis=1)
```

```python
combined.plot()
```

## Building a Backtesting system

```python
def predict(train, test, predictors, model):
    model.fit(train[predictors], train["Target"])
    preds = model.predict(test[predictors])
    preds = pd.Series(preds, index=test.index, name="Predictions")
    combined = pd.concat([test["Target"], preds], axis=1)
    return combined
```

```python
def backtest(data, model, predictors, start=2500, step=250):
    all_predictions = []
    
    for i in range(start, data.shape[0], step):
        train = data.iloc[0:i].copy()
        test = data.iloc[i:(i+step)].copy()
        predictions = predict(train, test, predictors, model)
        all_predictions.append(predictions)
    return pd.concat(all_predictions)
```

```python
predictions = backtest(nasdaq, model, predictors)
```

```python
predictions["Predictions"].value_counts()
```

```python
precision_score(predictions["Target"], predictions["Predictions"])
```

```python
predictions["Target"].value_counts() / predictions.shape[0]
```

## More predictors to the model

```python
horizons = [2, 5, 60, 250, 1000]
new_predictors = []

for horizon in horizons:
    rolling_averages = nasdaq.rolling(horizon).mean()
    
    ratio_column = f"Open_Ratio_{horizon}"
    nasdaq[ratio_column] = nasdaq["Open"] / rolling_averages["Open"]
    
    trend_column = f"Trend_{horizon}"
    nasdaq[trend_column] = nasdaq.shift(1).rolling(horizon).sum()["Target"]
    
    new_predictors += [ratio_column, trend_column]
```

```python
nasdaq = nasdaq.dropna()
```

```python
nasdaq
```

## Improving the model

```python
model = RandomForestClassifier(n_estimators = 200, min_samples_split = 50, random_state = 1)
```

```python
def predict(train, test, predictors, model):
    model.fit(train[predictors], train["Target"])
    preds = model.predict_proba(test[predictors])[:,1]
    preds[preds >= .6] = 1
    preds[preds < .6] = 0
    preds = pd.Series(preds, index=test.index, name="Predictions")
    combined = pd.concat([test["Target"], preds], axis=1)
    return combined
```

```python
predictions = backtest(nasdaq, model, new_predictors)
```

```python
predictions["Predictions"].value_counts()
```

```python
precision_score(predictions["Target"], predictions["Predictions"])
```
